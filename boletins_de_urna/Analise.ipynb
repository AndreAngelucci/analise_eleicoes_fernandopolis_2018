```python
import pandas as pd
pd.set_option('display.max_columns', 0)
pd.set_option('display.max_rows', 0)
```

```python
eleicoes = pd.read_csv('./eleicoes_final.csv', encoding='latin1')
```

# Analises
## Perguntas:
Qual a tendencia de voto de cada escola?

Qual escola votou mais no Bolsonaro/Haddad nos 2 turnos?

Como foram as mudanças de opinião no 2 turno?

```python
presidente = eleicoes[eleicoes['cargo'] == 'Presidente']
governador = eleicoes[eleicoes['cargo'] == 'Governador']
senador = eleicoes[eleicoes['cargo'] == 'Senador']
dep_federal = eleicoes[eleicoes['cargo'] == 'Deputado Federal']
dep_estadual = eleicoes[eleicoes['cargo'] == 'Deputado Estadual']
```

```python
# presidente
votos_por_turno_escola = pd.DataFrame(presidente.groupby(['turno', 'escola'])['voto_qtd'].sum())
votos_por_turno_escola_presidente = pd.DataFrame(presidente.groupby(['turno', 'escola', 'voto_nome'])['voto_qtd'].sum())
votos_por_turno_escola_presidente['percentual_votos'] = votos_por_turno_escola_presidente['voto_qtd'] / votos_por_turno_escola['voto_qtd'][votos_por_turno_escola_presidente.index] * 100
votos_por_turno_escola_presidente.to_csv('presidente_por_turno_escola.csv', encoding='latin1')
```

```python
# governador
votos_por_turno_escola = pd.DataFrame(governador.groupby(['turno', 'escola'])['voto_qtd'].sum())
votos_por_turno_escola_governador = pd.DataFrame(governador.groupby(['turno', 'escola', 'voto_nome'])['voto_qtd'].sum())
votos_por_turno_escola_governador['percentual_votos'] = votos_por_turno_escola_governador['voto_qtd'] / votos_por_turno_escola['voto_qtd'][votos_por_turno_escola_governador.index] * 100
votos_por_turno_escola_governador.to_csv('governador_por_turno_escola.csv', encoding='latin1')
```
